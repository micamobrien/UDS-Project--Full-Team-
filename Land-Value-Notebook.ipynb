```python
import requests
import json
import pandas as pd
import geopandas as gpd

ziplist = ['90017', '90057','90003','90012','90013']
#create empty df outside of loop
df_la = pd.DataFrame()
# for every zip in ziplist:
for x in ziplist: 
    r = requests.get('https://data.lacounty.gov/resource/9trm-uz8i.json?situszip5={}&$limit=1000000'.format(x))
    #load each url's info into a df
    d = pd.DataFrame(json.loads(r.text))
    #add that temporary df onto the empty list
    df_la = pd.concat([df_la, d], axis=0).reset_index(drop=True)
    
df_la
```
