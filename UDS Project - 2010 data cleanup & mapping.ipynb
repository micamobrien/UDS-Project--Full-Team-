```python
import json
import pandas as pd
import geopandas as gpd

gdf_2010 = gpd.read_file('/Users/jinglanlin/Desktop/School/01 UCLA/04 Spring 2022/UP 229 DS/Github/UDS-Project--Full-Team-/parcel_2010.geojson')
gdf_2010
```

```python
#query the data with land base year in 2010
gdf_2010.query('roll_landbaseyear == "2010"').roll_landbaseyear.count()
```

```python
#create a new dataframe subsetting all the parcels sold in 2010
gdf_2010_sold = gdf_2010[gdf_2010["roll_landbaseyear"] == '2010']
gdf_2010_sold
```

## Parcel geometry dataset

```python
#import parcel geometry data
import requests
import json


url = 'https://data.lacounty.gov/resource/rjze-ndj2.json?$where=within_box(the_geom, 34.1164, -118.3217, 34.01333, -118.2675)&$limit=100000'
r = requests.get(url)
parcel = pd.DataFrame(json.loads(r.text))
parcel
```

```python
#convert dataframe to geodataframe
parcelgdf = gpd.GeoDataFrame(
    parcel, geometry=gpd.points_from_xy(parcel.lon_coord, parcel.lat_coord, 
                                          crs='EPSG:4326'))
parcelgdf
```

```python
#drop columns for joining data
gdf_2010_sold = gdf_2010_sold.drop(columns=['index_right'])
gdf_2010_sold
```

```python
#Spatially join parcel data with geometry to data with land value
joinedgdf2010=gpd.sjoin(gdf_2010_sold.to_crs('EPSG:4326'),parcelgdf.to_crs('EPSG:4326'),how='left')
joinedgdf2010
```

```python
#merge land value data with parcel geometry data on ain
join2010 = gdf_2010_sold.merge(parcelgdf, how='left',on='ain')
join2010 = join2010.dropna()
join2010
```

```python
#counting numbers of ain in Westlake that were sold in 2010
join2010.query('neighborhood == "Westlake"').neighborhood.count()
```

```python
#convert datatypes
join2010['shape_area'] = pd.to_numeric(join2010['shape_area'])
join2010['roll_landvalue'] = pd.to_numeric(join2010['roll_landvalue'])
```

```python
#calculate the land value per square foot
join2010['landvalue_sqft']=join2010['roll_landvalue']/join2010['shape_area']
join2010
```

```python
#groubpy census tract and count the numbers of ain in each census tract
censusjoin2010 = join2010.groupby('CT20').ain.count()
censusjoin2010
```

```python
#convert the count to a pandas dataframe
df_census_n = pd.DataFrame(censusjoin2010)
df_census_n =df_census_n.rename(columns={'ain': 'n_parcel'})
df_census_n
```

```python
#find the sum land value of each census tract
censuslvsfjoin2010 = join2010.groupby('CT20').landvalue_sqft.sum()
censuslvsfjoin2010
```

```python
#convert it into a pandas dataframe
df_lvsf = pd.DataFrame(censuslvsfjoin2010)
df_lvsf
```

```python
#merge total land value of each census tract and number of parcel in each census tract for calculation
newjoin = df_lvsf.merge(df_census_n, how='left',on='CT20')
newjoin
```

```python
#calculate the average land value price per census tract
newjoin['avgprice_tract']=newjoin['landvalue_sqft']/newjoin['n_parcel']
newjoin
```

## Import average land price 2020 data

**creating in "Data Cleanup Land Value 2020" notebook**

```python
#read in the average land value 2020 data which we created in another notebook
avgprice_tract_2020 = pd.read_csv('/Users/jinglanlin/Desktop/School/01 UCLA/04 Spring 2022/UP 229 DS/Github/UDS-Project--Full-Team-/avgprice_tract_2020_new.csv')
avgprice_tract_2020
```

```python
#set CT20 as index
avgprice_tract_2020.set_index('CT20')
```

```python
#rename columns for 2020 data
avgprice_tract_2020 = avgprice_tract_2020.rename(columns={'landvalue_sqft': 'landvalue_sqft_2020',
                                                         'n_parcel': 'n_parcel_2020',
                                                         'avgprice_tract':'avgprice_tract_2020'})
avgprice_tract_2020
```

```python
#rename columns for 2010 data
avgprice_tract_2010 = newjoin.rename(columns={'landvalue_sqft': 'landvalue_sqft_2010',
                                                         'n_parcel': 'n_parcel_2010',
                                                         'avgprice_tract':'avgprice_tract_2010'})
avgprice_tract_2010
```

```python
#adjust the average price of 2010 to 2020 value; the CPI index is 1.19 using CPI Inflation Calculator on US Bureau of Labor Statistics website
avgprice_tract_2010['2010avgprice_adjust_to_2020']=avgprice_tract_2010['avgprice_tract_2010']*1.19
avgprice_tract_2010
```

```python
#reset index
avgprice_tract_2010.reset_index(inplace=True)
avgprice_tract_2010.info()
```

```python
#convert columns datatypes to float
avgprice_tract_2010['CT20'] = avgprice_tract_2010.CT20.astype(int)
avgprice_tract_2010['n_parcel_2010'] = avgprice_tract_2010.n_parcel_2010.astype(float)
avgprice_tract_2010.info()
```

```python
#convert columns datatypes to float again
avgprice_tract_2020['CT20'] = avgprice_tract_2020.CT20.astype(int)
avgprice_tract_2020['n_parcel_2020'] = avgprice_tract_2020.n_parcel_2020.astype(float)
avgprice_tract_2020.info()
```

```python
#joining the 2010 and 2020 average land price dataframes
finaljoin=avgprice_tract_2010.merge(avgprice_tract_2020, how='left',on='CT20')
finaljoin=finaljoin.dropna()
finaljoin
```

```python
#create two columns of percent change of land value from 2010 to 2020 with and without inflation
finaljoin['pct_change_wo_inflation'] = (finaljoin.loc[1:80,'avgprice_tract_2020'] - finaljoin.loc[1:80,'avgprice_tract_2010'])/finaljoin.loc[1:80,'avgprice_tract_2010']
finaljoin['pct_change_w_inflation']=(finaljoin.loc[1:80,'avgprice_tract_2020'] - finaljoin.loc[1:80,'2010avgprice_adjust_to_2020'])/finaljoin.loc[1:80,'2010avgprice_adjust_to_2020']
finaljoin
```

## MAPPING!

```python
#Bring in census tract data for geometry
census = gpd.read_file('/Users/jinglanlin/Desktop/School/01 UCLA/04 Spring 2022/UP 229 DS/Github/UDS-Project--Full-Team-/Data/Census_Tracts_2020.geojson')
census.head()
```

```python
#convert datatype to integer for merging
census['CT20']=census.CT20.astype(int)
```

```python
#merge the two dataframes for mapping
census_finaljoin=finaljoin.merge(census, on='CT20',how='left')
census_finaljoin
```

```python
#convert dataframe to geodataframe
import geopandas as gpd
gdf=gpd.GeoDataFrame(census_finaljoin,geometry='geometry')
```

```python
#drop columns for spatial join
gdf_2010= gdf_2010.drop(columns=['index_right'])
gdf_2010
```

```python
#spatially join land value dataset with gdf_2010 for neighborhood boundaries
gdf_map=gpd.sjoin(gdf.to_crs('EPSG:4326'),gdf_2010_sold.to_crs('EPSG:4326'),how='left')
gdf_map
```

```python
#drop the duplicates
gdf_map.drop_duplicates(subset='CT20_left', keep='first', inplace=True)
```

```python
#only keep the columns we need for mapping
finalgdf=gdf_map[['CT20_left','pct_change_wo_inflation', 'pct_change_w_inflation',
                     'geometry','center_lat','center_lon','neighborhood']]
finalgdf
```

```python
#convert this file to github for machine learning (random forest)
finalgdf.to_file('final_pctchange_landvalue.geojson', driver='GeoJSON') 
```
